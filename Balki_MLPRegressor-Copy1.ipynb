```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPRegressor

from datetime import datetime
now = datetime.now()
current_time = now.strftime("%H:%M:%S")
print("Current Time =", current_time)
# IMPORTING DATASET
# Zadnji stolpec je BA -> target, prvi stolpec je "Name" in je za odstrant
#df = pd.read_csv("pubchem_with_BA.csv", index_col=0) # Un Dataset parsed from PubChem
df = pd.read_csv("./data/LEK_data/clean.csv")
print(df.head())
# Normalizacija značilk, vsako značilko posebi
X_before_norm = df.iloc[:,:-1]
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X_before_norm)
X = pd.DataFrame(X, columns=X_before_norm.columns) # making it into DataFrame again

y = df.iloc[:, -1]
print("============FEATURES========")
for feature in X.columns:
    print(feature)
print()
print(X.head())
print(y.head())


average_scores = [] #used to keep track of best model

for hidden_layer_sizes in [(128, 64, 32), (1024, 512, 128, 32), (2048, 512, 64, 16)]:
    print("Hidden_layers_sizes: ", hidden_layer_sizes)
    
    for activation in ["identity", "logistic", "tanh", "relu"]:
        print("Activation: ", activation)
        
        for solver in ["lbfgs", "adam"]:
            print("Solver: ", solver)

            model = MLPRegressor(hidden_layer_sizes=hidden_layer_sizes, activation=activation, solver=solver)

            for split_size in [0.1, 0.15, 0.2, 0.25]:
                print("SPLIT SIZE: ", split_size)

                scores = []

                for i in range(10): # to average the score
                    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=split_size) # make train/test split, it is randomized
                    model.fit(X_train, y_train) # fit on the train
                    score = model.score(X_test, y_test) # get the score from test
                    scores.append(score)

                average_score = sum(scores)/len(scores)
                average_scores.append([average_score, hidden_layer_sizes, activation, solver, split_size])
            
best_result = max(average_scores, key=lambda i:i[0])
print("BEST RESULT: ", best_result[0], "\t Hidden layers: ", best_result[1], " \t Activation: ", best_result[2], "\t Solver: ", best_result[3], "\t Split_ratio: ", best_result[-1])
```
